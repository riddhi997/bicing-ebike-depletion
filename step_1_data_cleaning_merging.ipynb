```python
import pandas as pd
import numpy as np
import os
import glob
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')
```

```python
# 1. LOAD STATION METADATA FILES
print("Loading station metadata files...")
metadata_folder = "station_metadata_2019-2024"
metadata_files = glob.glob(os.path.join(metadata_folder, "*.csv"))

metadata_dict = {}
for file in metadata_files:
    # Extract year from filename (assuming format like "2019_INFO.csv")
    year = os.path.basename(file).split('_')[0]
    try:
        df = pd.read_csv(file)
        metadata_dict[year] = df
        print(f"Loaded {file} with shape: {df.shape}")
    except Exception as e:
        print(f"Error loading {file}: {e}")
```

```python
# 3. EXPLORE METADATA
print("\n" + "="*50)
print("METADATA EXPLORATION")
print("="*50)

# Check if we have metadata loaded
if metadata_dict:
    # Combine all metadata files
    all_metadata = pd.concat(metadata_dict.values(), ignore_index=True)
    
    # Remove duplicates based on station ID if exists
    if 'station_id' in all_metadata.columns:
        all_metadata = all_metadata.drop_duplicates(subset=['station_id'])
    
    print("\n1. Metadata Overview:")
    print(f"Total unique stations in metadata: {len(all_metadata)}")
    print(f"\nMetadata columns: {list(all_metadata.columns)}")
    
    print("\n2. Basic Metadata Info:")
    print(all_metadata.info())
    
    print("\n3. First few rows of metadata:")
    print(all_metadata.head())
    
    print("\n4. Metadata Statistics:")
    print(all_metadata.describe(include='all'))
    
    print("\n5. Missing values in metadata:")
    missing_metadata = all_metadata.isnull().sum()
    print(missing_metadata[missing_metadata > 0])
    
```

```python
#Drop Columns cross_street, planned_date, is_charging_station, x_ride_code_support, rental_uris, V1, is_valet_station, x_valet_station_details


all_metadata=all_metadata.drop(columns= ['cross_street', 'planned_date', 'is_charging_station', 'x_ride_code_support', 'rental_uris', 'V1', 'is_valet_station', 'x_valet_station_details'])
```

```python
# 2. LOAD MONTHLY STATION DATA FILES
print("\nLoading monthly station data files...")
data_folder = "stations_bicing_2019-2024"  

monthly_files = glob.glob(os.path.join(data_folder, "*_STATIONS_hourly_mean.csv"))

monthly_data_dict = {}
for file in monthly_files:
    # Extract year and month from filename (assuming format like "2019_03_STATIONS_hourly_mean.csv")
    filename = os.path.basename(file)
    year_month = filename.split('_')[0] + '_' + filename.split('_')[1]
    try:
        df = pd.read_csv(file)
        monthly_data_dict[year_month] = df
        print(f"Loaded {filename} with shape: {df.shape}")
    except Exception as e:
        print(f"Error loading {file}: {e}")
```

```python
# 4. EXPLORE MONTHLY DATA
print("\n" + "="*50)
print("MONTHLY DATA EXPLORATION")
print("="*50)

if monthly_data_dict:
    # Combine all monthly data
    all_monthly_data = []
    
    for period, df in monthly_data_dict.items():
        print(f"Processing: {period}")
        
        try:
            # Make a copy to avoid modifying the original
            df_copy = df.copy()
            
            # First, let's check what columns we have
            print(f"  Columns: {list(df_copy.columns)}")
            
            # Convert date column to datetime
            # Check for date column with different possible names
            date_col_name = None
            date_col_candidates = ['date', 'timestamp', 'datetime', 'time']
            
            for col in df_copy.columns:
                col_lower = col.lower()
                if any(candidate in col_lower for candidate in date_col_candidates):
                    date_col_name = col
                    print(f"  Found date column: {date_col_name}")
                    break
            
            if date_col_name:
                # Rename to standard 'date' column
                if date_col_name != 'date':
                    df_copy = df_copy.rename(columns={date_col_name: 'date'})
                
                # DEBUG: Show sample of date values before conversion
                print(f"  Sample date values (first 3): {df_copy['date'].head(3).tolist()}")
                
                # TRY DIFFERENT DATE PARSING METHODS
                original_date_dtype = df_copy['date'].dtype
                print(f"  Original date dtype: {original_date_dtype}")
                
                # Method 1: Try standard datetime parsing first
                df_copy['date_parsed'] = pd.to_datetime(df_copy['date'], errors='coerce')
                
                # Check if we got many NaT (not a time) values
                nat_count = df_copy['date_parsed'].isnull().sum()
                print(f"  NaT values after standard parsing: {nat_count}")
                
                # If we have many NaT values, try Unix timestamp parsing
                if nat_count > 0 and nat_count < len(df_copy):
                    print(f"  Trying Unix timestamp parsing for {nat_count} records...")
                    
                    # Check if values might be Unix timestamps (numeric or string numbers)
                    try:
                        # Convert to numeric if possible
                        numeric_dates = pd.to_numeric(df_copy['date'], errors='coerce')
                        
                        # Check if values are in Unix timestamp range
                        # Unix timestamps for 2019-2024 are around 1546300800 to 1704067200
                        unix_2019 = 1546300800  # 2019-01-01
                        unix_2024 = 1704067200  # 2024-01-01
                        
                        # Check how many values are in this range
                        valid_unix = numeric_dates.between(unix_2019, unix_2024).sum()
                        print(f"  Potential Unix timestamps (2019-2024 range): {valid_unix}")
                        
                        if valid_unix > 0:
                            # Try parsing as Unix timestamps (seconds)
                            df_copy.loc[df_copy['date_parsed'].isnull(), 'date_parsed'] = pd.to_datetime(
                                df_copy.loc[df_copy['date_parsed'].isnull(), 'date'].astype(float), 
                                unit='s', 
                                errors='coerce'
                            )
                            
                            # Also try as milliseconds if seconds didn't work
                            remaining_nat = df_copy['date_parsed'].isnull().sum()
                            if remaining_nat > 0:
                                df_copy.loc[df_copy['date_parsed'].isnull(), 'date_parsed'] = pd.to_datetime(
                                    df_copy.loc[df_copy['date_parsed'].isnull(), 'date'].astype(float), 
                                    unit='ms', 
                                    errors='coerce'
                                )
                    except:
                        print(f"  Could not parse as Unix timestamp")
                
                # Replace the original date column with parsed dates
                df_copy['date'] = df_copy['date_parsed']
                df_copy = df_copy.drop(columns=['date_parsed'])
                
                # Drop rows with invalid dates
                invalid_dates = df_copy['date'].isnull().sum()
                if invalid_dates > 0:
                    print(f"  Warning: Dropped {invalid_dates} rows with invalid dates")
                    df_copy = df_copy.dropna(subset=['date'])
                
                if len(df_copy) > 0:
                    # Extract year and month from ACTUAL DATE COLUMN (not filename)
                    df_copy['year'] = df_copy['date'].dt.year
                    df_copy['month'] = df_copy['date'].dt.month
                    df_copy['day'] = df_copy['date'].dt.day
                    df_copy['hour'] = df_copy['date'].dt.hour
                    
                    # Check year distribution in this file
                    year_counts = df_copy['year'].value_counts().head(5)
                    print(f"  Top years in this file: {dict(year_counts)}")
                    
                    # ===== KEY CHANGE: REMOVE 2020 NOW =====
                    records_before_2020 = len(df_copy)
                    # First filter to keep only 2019-2024
                    df_copy = df_copy[(df_copy['year'] >= 2019) & (df_copy['year'] <= 2024)]
                    
                    # THEN remove ALL of 2020
                    records_before_2020_removal = len(df_copy)
                    df_copy = df_copy[df_copy['year'] != 2020]
                    records_removed_2020 = records_before_2020_removal - len(df_copy)
                    
                    print(f"  Records before 2020 removal: {records_before_2020_removal}")
                    print(f"  Removed 2020 records: {records_removed_2020} ({(records_removed_2020/records_before_2020_removal*100):.1f}%)")
                    print(f"  Records after removing 2020: {len(df_copy)}")
                    
                    if len(df_copy) > 0:
                        # Add source filename for tracking
                        df_copy['source_file'] = period
                        
                        all_monthly_data.append(df_copy)
                        print(f"  Kept {len(df_copy)} records, date range: {df_copy['date'].min()} to {df_copy['date'].max()}")
                        print(f"  Years in this file: {sorted(df_copy['year'].unique())}")
                    else:
                        print(f"  Warning: No records left after removing 2020")
                else:
                    print(f"  Warning: No valid dates in this file")
            else:
                print(f"  Error: No date column found in this file")
                print(f"  Available columns: {list(df_copy.columns)}")
                
        except Exception as e:
            print(f"  Error processing {period}: {e}")
            import traceback
            traceback.print_exc()
            continue
    
    if all_monthly_data:
        # Combine all data
        combined_monthly = pd.concat(all_monthly_data, ignore_index=True)
        
        # Sort by date
        combined_monthly = combined_monthly.sort_values('date').reset_index(drop=True)
        
        # Add period column based on ACTUAL date (yyyy_mm format)
        combined_monthly['period'] = combined_monthly['date'].dt.strftime('%Y_%m')
        
        # Get proper date range
        min_date = combined_monthly['date'].min()
        max_date = combined_monthly['date'].max()
        
        print(f"\n" + "="*60)
        print("COMBINED DATA SUMMARY (2020 REMOVED)")
        print("="*60)
        print(f"Total files loaded: {len(all_monthly_data)}")
        print(f"Total records: {len(combined_monthly):,}")
        print(f"Date range: {min_date} to {max_date}")
        print(f"Time span: {(max_date - min_date).days} days")
        print(f"Number of unique stations: {combined_monthly['station_id'].nunique()}")
        print(f"Number of unique periods (YYYY_MM): {combined_monthly['period'].nunique()}")
        
        # Show data distribution by year
        print(f"\nRecords by year (2020 REMOVED):")
        yearly_counts = combined_monthly['year'].value_counts().sort_index()
        total_records = len(combined_monthly)
        for year, count in yearly_counts.items():
            print(f"  {year}: {count:,} records ({count/total_records*100:.1f}%)")
        
        # Verify 2020 is gone
        if 2020 in yearly_counts.index:
            print(f"\n⚠️  WARNING: 2020 data still present! {yearly_counts[2020]} records")
        else:
            print(f"\n✅ 2020 successfully removed!")
        
        # Check for date anomalies
        print(f"\nDate Quality Check:")
        
        # Check for dates before 2019 (should be none now)
        early_dates = combined_monthly[combined_monthly['year'] < 2019]
        if len(early_dates) > 0:
            print(f"  ⚠️  Warning: {len(early_dates):,} records before 2019 ({len(early_dates)/len(combined_monthly)*100:.2f}%)")
        else:
            print(f"  ✓ No records before 2019")
        
        # Check for future dates (after 2024)
        future_dates = combined_monthly[combined_monthly['year'] > 2024]
        if len(future_dates) > 0:
            print(f"  ⚠️  Warning: {len(future_dates):,} records after 2024 ({len(future_dates)/len(combined_monthly)*100:.2f}%)")
        else:
            print(f"  ✓ No records after 2024")
        
        # Check for duplicate timestamps per station
        duplicate_check = combined_monthly.duplicated(subset=['station_id', 'date']).sum()
        if duplicate_check > 0:
            print(f"  ⚠️  Warning: {duplicate_check:,} duplicate station-date combinations ({duplicate_check/len(combined_monthly)*100:.2f}%)")
        else:
            print(f"  ✓ No duplicate timestamps")
        
        # Check for missing values in key columns
        print(f"\nMissing Values Check:")
        key_columns = ['station_id', 'num_bikes_available', 'num_docks_available']
        for col in key_columns:
            if col in combined_monthly.columns:
                missing = combined_monthly[col].isnull().sum()
                if missing > 0:
                    print(f"  {col}: {missing:,} missing ({missing/len(combined_monthly)*100:.2f}%)")
                else:
                    print(f"  ✓ {col}: no missing values")
        
        # Display basic statistics for numeric columns
        print(f"\nBasic Statistics for Key Variables:")
        numeric_cols = ['num_bikes_available', 'num_docks_available']
        
        # Add bike type columns if they exist
        bike_type_cols = ['num_bikes_available_types.mechanical', 'num_bikes_available_types.ebike']
        for col in bike_type_cols:
            if col in combined_monthly.columns:
                numeric_cols.append(col)
        
        for col in numeric_cols:
            if col in combined_monthly.columns:
                print(f"\n{col}:")
                print(f"  Mean: {combined_monthly[col].mean():.2f}")
                print(f"  Median: {combined_monthly[col].median():.2f}")
                print(f"  Std: {combined_monthly[col].std():.2f}")
                print(f"  Min: {combined_monthly[col].min():.2f}")
                print(f"  Max: {combined_monthly[col].max():.2f}")
        
        # Show first few rows
        print(f"\nFirst 3 rows of cleaned data:")
        print(combined_monthly.head(3))
        
        # Show column information
        print(f"\nColumns in cleaned data ({len(combined_monthly.columns)} total):")
        for i, col in enumerate(combined_monthly.columns, 1):
            print(f"  {i:2d}. {col} ({combined_monthly[col].dtype})")
    
    else:
        print("No monthly data files were successfully processed")
else:
    print("No monthly data files were loaded")

print("\n" + "="*50)
print("MONTHLY DATA EXPLORATION AND CLEANING COMPLETE")
print("="*50)
```

```python
print("\n" + "="*70)
print("MERGE MONTHLY DATA WITH METADATA")
print("="*70)

# Check if we have both datasets
if 'combined_monthly' not in locals() and 'combined_monthly' not in globals():
    print("Error: 'combined_monthly' DataFrame not found!")
elif 'all_metadata' not in locals() and 'all_metadata' not in globals():
    print("Error: 'all_metadata' DataFrame not found!")
else:
    print(f"Monthly data shape: {combined_monthly.shape}")
    print(f"Metadata shape: {all_metadata.shape}")
```

```python
# 1. PREPARE MONTHLY DATA
print("\n" + "="*70)
print("1. PREPARING MONTHLY DATA")
print("="*70)

# Make a copy to avoid modifying original
monthly_data = combined_monthly.copy()

# Calculate capacity from actual data (bikes + docks)
monthly_data['capacity'] = monthly_data['num_bikes_available'] + monthly_data['num_docks_available']

# Select and rename columns from monthly data
monthly_columns_to_keep = {
    'date': 'date',
    'station_id': 'station_id',
    'num_bikes_available': 'bikes_available',
    'num_bikes_available_types.mechanical': 'mechanical_bikes',
    'num_bikes_available_types.ebike': 'ebikes',
    'num_docks_available': 'docks_available',
    'year': 'year',
    'month': 'month',
    'day': 'day',
    'hour': 'hour',
    'capacity': 'capacity'  # New calculated column
}

# Filter to only keep columns that exist in the dataframe
existing_columns = [col for col in monthly_columns_to_keep.keys() if col in monthly_data.columns]
monthly_data_clean = monthly_data[existing_columns].copy()

# Rename columns
monthly_data_clean = monthly_data_clean.rename(columns=monthly_columns_to_keep)

print(f"Monthly data columns kept: {list(monthly_data_clean.columns)}")
print(f"Monthly data shape after cleaning: {monthly_data_clean.shape}")
```

```python
# 2. PREPARE METADATA
print("\n" + "="*70)
print("2. PREPARING METADATA")
print("="*70)

# Make a copy to avoid modifying original
metadata = all_metadata.copy()

# Select and rename columns from metadata
metadata_columns_to_keep = {
    'station_id': 'station_id',
    'lat': 'latitude',
    'lon': 'longitude',
    'altitude': 'altitude'
}

# Filter to only keep columns that exist in the dataframe
existing_metadata_cols = [col for col in metadata_columns_to_keep.keys() if col in metadata.columns]
metadata_clean = metadata[existing_metadata_cols].copy()

# Rename columns
metadata_clean = metadata_clean.rename(columns=metadata_columns_to_keep)

# Remove duplicates based on station_id (keep first occurrence)
metadata_clean = metadata_clean.drop_duplicates(subset=['station_id'])

print(f"Metadata columns kept: {list(metadata_clean.columns)}")
print(f"Metadata shape after cleaning: {metadata_clean.shape}")
print(f"Unique stations in metadata: {metadata_clean['station_id'].nunique()}")
```

```python
# 3. MERGE DATASETS
print("\n" + "="*70)
print("3. MERGING DATASETS")
print("="*70)

# Perform left join (keep all monthly records)
merged_data = pd.merge(
    monthly_data_clean,
    metadata_clean,
    on='station_id',
    how='left',
    validate='many_to_one'  # Ensures each monthly record matches at most one station
)

print(f"Merged data shape: {merged_data.shape}")
print(f"Columns in merged data: {list(merged_data.columns)}")

```

```python
# 4. DATA QUALITY CHECKS AFTER MERGING
print("\n" + "="*70)
print("4. DATA QUALITY CHECKS")
print("="*70)

# Check for stations in monthly data that don't have metadata
stations_without_metadata = merged_data[merged_data['latitude'].isnull()]['station_id'].unique()
if len(stations_without_metadata) > 0:
    print(f"⚠️  Warning: {len(stations_without_metadata)} stations in monthly data have no metadata:")
    for station in stations_without_metadata[:10]:  # Show first 10
        print(f"  Station {station}")
    if len(stations_without_metadata) > 10:
        print(f"  ... and {len(stations_without_metadata) - 10} more")
    
    # Count records without metadata
    records_without_metadata = merged_data['latitude'].isnull().sum()
    print(f"  Total records without location data: {records_without_metadata:,} ({records_without_metadata/len(merged_data)*100:.2f}%)")
else:
    print("✓ All stations have metadata")

# Check capacity calculation sanity
print(f"\nCapacity calculation check:")
print(f"  Average capacity: {merged_data['capacity'].mean():.2f}")
print(f"  Minimum capacity: {merged_data['capacity'].min():.2f}")
print(f"  Maximum capacity: {merged_data['capacity'].max():.2f}")

# Check for impossible values (negative or zero capacity)
invalid_capacity = merged_data[merged_data['capacity'] <= 0]
if len(invalid_capacity) > 0:
    print(f"⚠️  Warning: {len(invalid_capacity):,} records have invalid capacity (<= 0)")
    print(f"  These will be removed for analysis")

# Check if bikes exceed capacity
bikes_exceed_capacity = merged_data[merged_data['bikes_available'] > merged_data['capacity']]
if len(bikes_exceed_capacity) > 0:
    print(f"⚠️  Warning: {len(bikes_exceed_capacity):,} records have more bikes than capacity")
    print(f"  Max excess: {(bikes_exceed_capacity['bikes_available'] - bikes_exceed_capacity['capacity']).max():.2f}")
```

```python
# 5. ADD DERIVED FEATURES
print("\n" + "="*70)
print("5. ADDING DERIVED FEATURES")
print("="*70)

# Add temporal features
merged_data['date'] = pd.to_datetime(merged_data['date'])
merged_data['day_of_week'] = merged_data['date'].dt.dayofweek  # Monday=0, Sunday=6
merged_data['day_name'] = merged_data['date'].dt.day_name()
merged_data['is_weekend'] = merged_data['day_of_week'].isin([5, 6]).astype(int)

# Add time of day categories
def categorize_time_of_day(hour):
    if 0 <= hour < 6:
        return 'night'
    elif 6 <= hour < 10:
        return 'morning_peak'
    elif 10 <= hour < 16:
        return 'midday'
    elif 16 <= hour < 20:
        return 'evening_peak'
    else:
        return 'evening'

merged_data['time_of_day'] = merged_data['hour'].apply(categorize_time_of_day)

# Add season
def get_season(month):
    if month in [12, 1, 2]:
        return 'winter'
    elif month in [3, 4, 5]:
        return 'spring'
    elif month in [6, 7, 8]:
        return 'summer'
    else:
        return 'autumn'

merged_data['season'] = merged_data['month'].apply(get_season)

# Add utilization metrics
merged_data['utilization_rate'] = merged_data['bikes_available'] / merged_data['capacity'].replace(0, np.nan)
merged_data['available_dock_rate'] = merged_data['docks_available'] / merged_data['capacity'].replace(0, np.nan)

# Add bike type ratios (if columns exist)
if 'mechanical_bikes' in merged_data.columns and 'ebikes' in merged_data.columns:
    merged_data['mechanical_ratio'] = merged_data['mechanical_bikes'] / merged_data['bikes_available'].replace(0, np.nan)
    merged_data['ebike_ratio'] = merged_data['ebikes'] / merged_data['bikes_available'].replace(0, np.nan)

print(f"Added features: day_of_week, is_weekend, time_of_day, season, utilization_rate, available_dock_rate")
if 'mechanical_ratio' in merged_data.columns:
    print(f"Added bike type ratios: mechanical_ratio, ebike_ratio")
```

```python
# 6. FINAL CLEANUP
print("\n" + "="*70)
print("6. FINAL CLEANUP")
print("="*70)

# Remove records with invalid capacity
original_count = len(merged_data)
merged_data = merged_data[merged_data['capacity'] > 0]
final_count = len(merged_data)

if original_count > final_count:
    removed = original_count - final_count
    print(f"Removed {removed:,} records with invalid capacity ({removed/original_count*100:.2f}%)")

# Cap utilization rate at 1.0 (in case bikes exceeded capacity)
merged_data['utilization_rate'] = merged_data['utilization_rate'].clip(0, 1)
merged_data['available_dock_rate'] = merged_data['available_dock_rate'].clip(0, 1)
```
